```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("performance-test.csv")

colors = {'Single Thread': '#929292', 'Multi Thread': '#929292', 'Multiplexing': '#89a5ea'}

for method in ['Single Thread', 'Multi Thread', 'Multiplexing']:
    plt.plot(df.index, df[method], marker='o',markersize=1,color=colors[method], label=method)

plt.xlabel("Test Count")
plt.ylabel("Runtime (ms)")
plt.title("Performance Test")
plt.legend()
plt.grid(True)


plt.show()
```
